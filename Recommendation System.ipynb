# Data Preprocessing

```python
df = pd.read_csv(r"C:\Users\reddy\Downloads\data assingnment files\Assingnment files\Recommendation System\anime.csv")
```

```python
df.head()
```

```python
df1 = df.pivot_table(index='name', columns='anime_id', values=['genre', 'type', 'episodes', 'rating', 'members'],aggfunc='first')
```

## Handle missing values

```python
df1.fillna(0,inplace=True)
```

```python
df1
```

# Feature Extraction

```python
from sklearn.metrics.pairwise import cosine_similarity,euclidean_distances 
```

```python
df_numeric = df.select_dtypes(include=[float, int])
df_numeric.fillna(0,inplace=True)
```

```python
cosine_sim = cosine_similarity(df_numeric)
cosine_sim
```

# Recommendation System

```python
def recommended_movie(movie_name):
    if movie_name in df1.index:
       index =np.where(movie_name==df1.index)[0][0]
       similar_movie= sorted(list(enumerate(cosine_sim[index])), key=lambda x: x[1], reverse=True)[1:6]
       print(f'Recommended Movies of {movie_name}')
       print('-'*12)
       for movie in similar_movie:
          print(df1.index[movie [0]])
    else:
       print('Movie is not in the list')
```

```python
recommended_movie('Kimi no Na wa.')
```

## Evaluate the recommendation system using appropriate metrics such as precision, recall, and F1-score.

```python
recommended_ids = df['anime_id'].tolist()
actual_ids = df['anime_id'].tolist()
```

```python
precision = precision_score(actual_ids, recommended_ids, average='micro') # Use 'micro' for overall precision
recall = recall_score(actual_ids, recommended_ids, average='micro') # Use 'micro' for overall recall
f1 = f1_score(actual_ids, recommended_ids, average='micro') # Use 'micro' for overall F1-score
```
